```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
excel_url = './data.xlsx'
df = pd.read_excel(excel_url, sheet_name = 4)
condition = (df['접종일'] >= "2021.01.01") & (df['접종일'] <= "2021.09.31") #조건식 작성
df=df[condition]
df
```

```python
import calendar
x_data = ["21.01","21.02","21.03","21.04","21.05","21.06","21.07","21.08","21.09"]
y_data = []
for i in range(1,10):
    day = calendar.monthrange(2021, i)[1]
    condition = (df['접종일'] == "2021."+"0"+str(i)+"."+str(day))
    if(len(df[condition]) == 0):
        y_data.append(0)
    else:
        y_data.append(df[condition]['2차접종률(%)'])
            
x_data2 = [202101,202102,202103,202104,202105,202106,202107,202108,202109]

plt.ylim([10, 100])
plt.bar(x_data2,y_data)
```
